```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(df['genres'])
print(feature_vectors)
```

### cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(feature_vectors)
similarity
```

#### creating a list with all the movies given in the dataset

```python
list_of_titles=df['title'].tolist()
list_of_titles
```

```python
movie_name=input("Enter movie name:")
```

```python
import difflib
find_close_match=difflib.get_close_matches(movie_name,list_of_titles)
find_close_match
```

```python
close_match=find_close_match[0]
```

```python
dir(difflib)
```

```python
index_of_movies=df[df.title==close_match]['index'].values[0]
index_of_movies
```

```python
similarity_score=list(enumerate(similarity[index_of_movies]))
similarity_score
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
sorted_similar_movies
```

# print the name of similar movies based on index

```python
print("movie suggested for you: \n")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if (i<=10):
        print(i,'.',title_from_index)
    i+=1
```
